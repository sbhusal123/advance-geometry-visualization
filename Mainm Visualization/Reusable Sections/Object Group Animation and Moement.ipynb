```python
from manim import *

class GroupAnimationExample(Scene):
    def construct(self):
        # 1️⃣ Create individual pieces
        a = MathTex("a")
        plus = MathTex("+")
        b = MathTex("b")
        equals = MathTex("=")
        c = MathTex("c")
        
        # 2️⃣ Arrange them horizontally
        formula_group = VGroup(a, plus, b, equals, c).arrange(RIGHT, buff=0.5)
        
        # 3️⃣ Add to scene
        self.play(Write(formula_group))
        self.wait(1)
        
        # 4️⃣ Animate part of the group
        self.play(a.animate.set_color(RED), b.animate.set_color(BLUE))
        self.wait(1)
        
        # 5️⃣ Animate the whole group together
        self.play(formula_group.animate.shift(UP*2))
        self.wait(1)
        
        # 6️⃣ Scale the group
        self.play(formula_group.animate.scale(1.5))
        self.wait(2)

%manim -ql -v ERROR GroupAnimationExample
```
