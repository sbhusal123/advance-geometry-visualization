```python
from manim import *

class ConeOrthographicViews(Scene):
    def construct(self):
        # ---------------- Parameters ----------------
        cone_radius = 20  # radius of cone base
        cone_height = 40  # height of cone
        hole_diameter = 5  # diameter of hole
        hole_radius = hole_diameter / 2
        
        # ---------------- Top View ----------------
        top_view_cone = Circle(radius=cone_radius, color=BLUE)
        top_view_hole = Circle(radius=hole_radius, color=RED).shift(LEFT * cone_radius/2)
        top_view_group = VGroup(top_view_cone, top_view_hole).to_corner(UL)
        top_label = Text("Top View", font_size=24).next_to(top_view_group, UP)
        
        # ---------------- Left View ----------------
        left_view_cone = Triangle().scale(cone_height / 2).rotate(-PI/2)  # Simple triangle side view
        left_view_hole = Circle(radius=hole_radius, color=RED).shift(LEFT * cone_height/4)
        left_view_group = VGroup(left_view_cone, left_view_hole).to_corner(DL)
        left_label = Text("Left View", font_size=24).next_to(left_view_group, LEFT)
        
        # ---------------- Right View ----------------
        right_view_cone = Triangle().scale(cone_height / 2).rotate(PI/2)
        right_view_group = VGroup(right_view_cone).to_corner(DR)
        right_label = Text("Right View", font_size=24).next_to(right_view_group, RIGHT)
        
        # ---------------- Add all ----------------
        self.add(top_view_group, top_label)
        self.add(left_view_group, left_label)
        self.add(right_view_group, right_label)

%manim -ql -v ERROR ConeOrthographicViews
```
